```python

def monte_carlo_simulation(start_price, volatility, num_simulations, num_days):
    results = np.zeros((num_simulations, num_days))

    for i in range(num_simulations):
        daily_returns = np.random.normal(0, volatility, num_days)  # روزانه تغییرات قیمت
        price_path = start_price * np.exp(np.cumsum(daily_returns))  # شبیه‌سازی قیمت در هر روز
        results[i] = price_path

    return results

# پارامترهای شبیه‌سازی
start_price = 10000  # قیمت اولیه
volatility = 0.05  # نوسانات روزانه (۵%)
num_simulations = 1000  # تعداد شبیه‌سازی‌ها
num_days = 30  # تعداد روزها

# اجرای شبیه‌سازی مونت کارلو
simulated_prices = monte_carlo_simulation(start_price, volatility, num_simulations, num_days)

# رسم نتایج
plt.figure(figsize=(10, 6))
plt.plot(simulated_prices.T, color='blue', alpha=0.1)  # نمایش نتایج شبیه‌سازی‌ها
plt.title('Monte Carlo Simulation for Crypto Price Prediction')
plt.xlabel('Days')
plt.ylabel('Price')
plt.show()
```

```python
# محاسبه VaR و CVaR
def calculate_var_and_cvar(simulated_prices, confidence_level=0.05):
    # محاسبه تغییرات قیمت در آخرین روز
    end_prices = simulated_prices[:, -1]
    returns = end_prices / start_price - 1

    # محاسبه VaR
    var = np.percentile(returns, 100 * confidence_level)

    # محاسبه CVaR
    cvar = returns[returns <= var].mean()

    return var, cvar

# محاسبه VaR و CVaR
var, cvar = calculate_var_and_cvar(simulated_prices)
print(f"Value at Risk (VaR) at 5% level: {var * 100:.2f}%")
print(f"Conditional Value at Risk (CVaR) at 5% level: {cvar * 100:.2f}%")
```
